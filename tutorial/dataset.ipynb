# Tensorflow Dataset

```python
import tensorflow as tf
import numpy as np
```

```python
tf.__version__
```

```python
def next_sample(data_iter):
    session = tf.get_default_session()
    if not session:
        return None
    session.run(data_iter.initializer)
    a = session.run(data_iter.get_next())
    return a
```

```python
data = tf.data.Dataset.from_tensors(np.random.random(size = [3, 4]))
data
```

```python
data = data.repeat(4) # repeat 4 times, if data = data.repeat() means repeat forever.
```

```python
with tf.Session() as sess:
    data_iter = data.make_initializable_iterator()
    sample = next_sample(data_iter)
    print(sample)
```

```python
data = data.batch(4)
```

```python
with tf.Session() as sess:
    data_iter = data.make_initializable_iterator()
    sample = next_sample(data_iter)
    print(sample)
```

```python
data = tf.data.Dataset.from_tensor_slices(np.random.random(size = [3, 4]))
data
```

```python
data = data.batch(3)
data = data.repeat(10)
```

```python
with tf.Session() as sess:
    data_iter = data.make_initializable_iterator()
    sample = next_sample(data_iter)
    print(sample)
```
